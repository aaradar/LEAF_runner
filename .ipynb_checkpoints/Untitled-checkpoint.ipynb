```python
import xarray as xr
import numpy as np
import pandas as pd

# Create example data
times = pd.date_range('2023-01-01', periods=5, freq='D')
x_coords = range(4)
y_coords = range(4)
data1 = np.random.rand(5, 4, 4)
data2 = np.random.rand(5, 4, 4)

# Create a Dataset
ds = xr.Dataset({
    'variable1': (['time', 'x', 'y'], data1),
    'variable2': (['time', 'x', 'y'], data2)
}, coords={'time': times, 'x': x_coords, 'y': y_coords})

# Print the dataset
print("Original Dataset:")
print(ds)

# Find the indices of the maximum values along the 'time' dimension for 'variable1'
max_indices = ds['variable1'].argmax(dim='time')

# Use these indices to select the corresponding pixels from both variables
selected_data1 = ds['variable1'].isel(time=max_indices)
selected_data2 = ds['variable2'].isel(time=max_indices)

# Create a new Dataset with the selected data
selected_ds = xr.Dataset({
    'variable1_max': selected_data1,
    'variable2_corresponding': selected_data2
})

# Print the new dataset
print("\n\n\nSelected Dataset with Maximum Values:")
print(selected_ds)
```
