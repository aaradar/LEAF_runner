```python
ottawa_region = {
    'type': 'Polygon',
    'coordinates': [
       [
         [-76.120,45.184], 
         [-75.383,45.171],
         [-75.390,45.564], 
         [-76.105,45.568], 
         [-76.120,45.184]
       ]
    ]
}

catalog_name = "https://earth-search.aws.element84.com/v1"
coll_name    = "sentinel-2-l2a"
start_str    = '2022-07-01'
end_str      = '2022-08-31'

#%%time
stac_IC = stacM.get_STAC_IC(catalog_name, coll_name, ottawa_region, 'EPSG:3979', start_str, end_str)

#stac_IC.red.plot.imshow(col="time")
display(stac_IC)
```

```python
_ = (
    stac_IC   #.isel(time=0)
    .to_array("band")
    .plot.imshow(
        col="band",
        size=4,
        vmin=0,
        vmax=4000,
    )
)
```
